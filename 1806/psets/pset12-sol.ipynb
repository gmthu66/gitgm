### Solution

You can just copy right out of the book to find out which directions to take, but to emphasize understanding here is a detailed and concise (compared to the textbook) explanation:

We are asked to pick out two directions in 4d that explain as much of the variation of the data as we can.  What this means is the following.  A direction in 4d is a unit vector $v = (v_1, v_2, v_3, v_4) \in \mathbb{R}^4$.  Given such a direction, we can take the matrix product $Av$, which gives a vector in $\mathbb{R}^{150}$, because $A$ is $150 \times 4$ and $v$ is $4 \times 1$.  Recall from our interpretations of matrix multiplication that this product $Av$ is actually the following combination of vectors: $$Av = v_1\alpha_1 + v_2\alpha_2 + v_3\alpha_3 + v_4\alpha_4$$ where $\alpha_1, \alpha_2, \alpha_3, \alpha_4 \in \mathbb{R}^{150}$ are the 4 columns of $A$.  So, $Av$ is the vector, of same length as the number of flowers, that records for each flower the linear combination of the 4 measurements for that flower each taken with coefficients $v_1, v_2, v_3, v_4$, respectively.  We can think about this vector $Av$ as giving the "component of our data in the $v$ direction".  Important examples to get a feel for this are the following.  First, if $v = e_1 = (1, 0, 0, 0)$, then $Av$ is just the first column of $A$, i.e. it's just the first measurement taken for each flower.  Similarly, if $v = e_3 = (0, 0, 1, 0)$, then $Av$ is just the third measurement for each flower.  If every single one of the flower measurement vectors in $\mathbb{R}^4$ happened to be orthogonal to $v$, then the "component of our data in the $v$ direction" should be 0, and indeed $Av = 0$ in this case.

We're interested in the opposite thing, i.e. in directions $v$ so that $Av$ retains as much information as possible.  This should mean that the combination measurements $Av$ in the $v$ direction should be collectively large over the 150 flowers, which should mean that $\Vert Av\Vert$ should be as large as possible.  The SVD $A = U\Sigma V^T$ is exactly the tool we need to decide which direction $v$ is best to take to make $Av$ large; the SVD says that $Av = U\Sigma V^Tv$, and as $U$ and $V$ are orthogonal matrices they don't increase the size of vectors, so all of the scaling of size comes from the singular values appearing in $\Sigma$.  It's not so hard to see that the unit vector $w \in \mathbb{R}^4$ that makes $\Vert \Sigma w \Vert$ as large as possible is $w = e_1 = (1, 0, 0, 0)$ (or its negative), because the singular values appear along the diagonal of $\Sigma$ in decreasing order of magnitude.  It follows that the $v \in \mathbb{R}^4$ that makes $\Vert Av\Vert = \Vert U\Sigma V^Tv\Vert$ as big as possible should be the $v$ such that $V^Tv = e_1$, i.e. $v = V^{-1}e_1 = Ve_1$, i.e. $v$ should be the first column $V[:,1]$ of $V$.  Having found this, we can now look for the second best direction $v$ to take, i.e. the best direction $v$ to take *among those directions orthogonal to* $V[:,1]$.  By the same reasoning, if $v$ is such a direction, then $w = V^Tv$ should make $\Vert\Sigma w\Vert$ as large as possible among those $w$ orthogonal to $V^TV[:,1] = e_1 = (1, 0, 0, 0)$, so $w$ should be $w = e_2 = (0, 1, 0, 0)$.  So $e_2 = w = V^Tv$, so now $v = Ve_2$, i.e. the second direction to take should be the second column of $V$.

So, in the Julia code we want to take `x1 = A*V[:,1]` and `x2 = A*V[:,2]`.   Alternatively, since $AV = U\Sigma$, these vectors are in the same directions as the first two *left* singular vectors ($Av = \sigma u$), i.e. we could also use `x1 = U[:,1]` and `x2 = U[:,2]` .

## Problem 2 (7+7+7 points)

From the lecture 31 notes, recall that if $Ax_1 = \lambda_1 x_1$ is the only linearly independent eigenvector for an eigenvalue of multiplicity 2 (a double root), then we can also find a Jordan vector with $Aj_1 = \lambda_1 j_1 + x_1$, typically chosen $\perp x_1$ ($j_1^H x_1 = 0$).   If $f(A)$ is some function of the matrix $A$, we also said that $f(A)x_1 = f(\lambda_1)x_1$ while $f(A) j_1 = f(\lambda_1) j_1 + f'(\lambda_1) x_1$.

**(a)** What is $y = A^{-1} j_1$, i.e. applying the rule for $f(A)=A^{-1}$?  Check that you get $Ay = j_1$ from the rules for multiplying by $A$.

**(b)** If this defective $A$ is similar to a matrix $B = SAS^{-1}$, give an eigenvector and the corresponding Jordan vector for $B$.

**(c)** For this defective $A$, suppose that you know that all the *other* eigenvalues $\lambda_2,\ldots$ are "simple" (i.e. multiplicity 1, i.e. single roots, i.e. $1\times 1$ Jordan blocks, i.e. they just have ordinary eigenvectors).  Under what condition on the eigenvalues $\lambda_1,\lambda_2,\ldots$ must $\frac{dx}{dt}=Ax$ have solutions $x(t)$ that approach a nonzero constant vector for $t\to\infty$, for most $x(0)$?

### Solution

**(a)**  This is the case $f(x) = x^{-1}$.  We have $f'(x) = -x^{-2}$ in this case, so we get $$y = A^{-1}j_1 = f(A)j_1 = f(\lambda_1)j_1 + f'(\lambda_1)x_1 = \lambda_1^{-1}j_1 - \lambda_1^{-2}x_1.$$  To check that $Ay = j_1$, i.e. so that writing "$y = A^{-1}j_1$" was a reasonable thing, we compute $$Ay = A(\lambda_1^{-1}j_1 - \lambda_1^{-2}x_1) = \lambda_1^{-1}Aj_1 - \lambda_1^{-2}Ax_1 = \lambda_1^{-1}(\lambda_1j_1 + x_1) - \lambda_1^{-2}\lambda_1x_1 = j_1.$$

**(b)**  The matrix $B = SAS^{-1}$ is obtained from the matrix $A$ by a change of coordinates specified by the matrix $S$.  Intuitively, this means that whatever the matrix $A$ does to vectors $x, y, z$ the matrix $B$ does similar things to the vectors $Sx, Sy, Sz$.  So, if $x_1$ is an eigenvector for $A$ with eigenvalue $\lambda_1$ and $j_1$ is an associated Jordan vector, then $Sx_1$ is an eigenvector for $B$ with eigenvalue $\lambda_1$ and with associated Jordan vector $Sj_1$.  To see that $Sx_1$ is indeed an eigenvector for $B$ with eigenvalue $1$, we compute: $$BSx_1 = SAS^{-1}Sx_1 = SAx_1 = S\lambda_1x_1 = \lambda_1(Sx_1)$$ so indeed $Sx_1$ is an eigenvector for $B$ with eigenvalue $\lambda_1$.  To check that $Sj_1$ is an associated Jordan vector, we compute $$BSj_1 = SAS^{-1}Sj_1 = SAj_1 = S(\lambda_1j_1 + x_1) = \lambda_1Sj_1 + Sx_1$$ as needed.

Recall that the Jordan vectors are not unique (we could add any multiple of $x_1$); we sometimes choose $j_1$ to be orthogonal to $x_1$.   However, $Sj_1$ is *not* in general orthogonal to $Sx_1$ unless $S$ happens to be unitary.   If we wanted to enforce this property (which is *not required* in this problem), we would have to use Gram–Schmidt, and choose the Jordan vector:
$$
Sj_1 - x_1 \frac{x_1^T S^T S j_1}{x_1^T S^T S x_1}
$$

**(c)**  To not have clashing notation between the initial condition $x(0)$ and solution $x(t)$ and the eigenvector $x_1$, I'll use $v$'s for eigenvectors here.  The conditions on the matrix $A$ means that there is a basis $v_1, j_1, v_2, v_3, ...$ (for $\mathbb{R}^m$ if $A$ is $m \times m$, let's say) such that $v_1, j_1$ are as in the problem statement ($v_1$ is an eigenvector for $A$ with eigenvalue $\lambda_1$ and $j_1$ is an associated Jordan vector) and so that $v_2$ is an eigenvector with eigenvalue $\lambda_2$, $v_3$ is an eigenvector with eigenvalue $\lambda_3$, etc. We're interested in vector-valued functions $x(t)$ that are solutions to the differential equation $x'(t) = Ax(t)$. Multiplying by $A$, we get $AA^TAv = A\lambda v$, so that $(AA^T)(Av) = \lambda(Av).$  So, as long as $\lambda Av \neq 0$, then this means that $\boxed{Av}$ is also an eigenvector of $AA^T$.  So when is $\lambda Av \neq 0$?  Well, we need both that $\lambda \neq 0$ and that $Av \neq 0$.  But in fact the latter follows from the former.  Indeed, if $\lambda \neq 0$, then that $v$ is an eigenvector of $A^TA$ with eigenvalue $\lambda$ means that $A^TAv = \lambda v \neq 0$, so $v$ is not in the nullspace of $A^TA$. **(b)**  Well, we have $$B = Q_mAQ_n^T = Q_mU\Sigma V^TQ_n^T = (Q_mU)\Sigma(VQ_n)^T.$$  But the product of two orthogonal matrices is still orthogonal (why?), so $Q_mU$ and $VQ_n$ are orthogonal matrices, and $\Sigma$ still only has zero entries on the diagonal, so this is exactly a singular value decomposition of the matrix $B$. **(c)**  The code is run below.
